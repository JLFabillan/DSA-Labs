```python
class Node():
    def __init__(self, data=None, next=None):
        self.data = data
        self.next = None

class LinkedList:
    def __init__ (self):
        self.head = None
        self.tail = None

    def insert_at_beginning(self, data):
        new_node = Node(data)
        if self.head:
            new_node.next = self.head
            self.head = new_node
        else:
            self.tail = new_node
            self.head = new_node

    def insert_at_end(self, data):
        new_node = Node(data)
        if self.head:
            self.tail.next = new_node
            self.tail = new_node
        else:
            self.head = new_node
            self.tail = new_node

    def search(self, data):
        current_node = self.head
        while current_node:
            if current_node.data == data:
                return True
            else:
                current_node = current_node.next
        return False

# a. remove_beginning
    def remove_beginning(self):
        if not self.head:
            return None
        removed_data = self.head.data
        self.head = self.head.next
        # if list becomes empty
        if self.head is None:
            self.tail = None
        return removed_data

# b. remove_at_end
    def remove_at_end(self):
        if not self.head:
            return None
        # If only one node exists
        if self.head == self.tail:
            removed_data = self.head.data
            self.head = None
            self.tail = None
            return removed_data
        # Traverse to the second last node
        current = self.head
        while current.next != self.tail:
            current = current.next
        removed_data = self.tail.data
        current.next = None
        self.tail = current
        return removed_data

# c. remove_at
    def remove_at(self,data) :
        if not self.head:
            return None

        # If the node to remove is the head
        if self.head.data == data:
            return self.remove_beginning()

        current = self.head
        while current.next:
            if current.next.data == data:
                removed_data = current.next.data
                current.next = current.next.next

                # If removed node was tail update tail
                if current.next is None:
                    self.tail = current

                return removed_data

            current = current.next
        return None

# display the list
    def display(self):
        current = self.head
        while current is not None:
            print(current.data, end=" -> ")
            current = current.next
        print("None")


# Test
sushi_preparation = LinkedList()
sushi_preparation.insert_at_end("prepare")
sushi_preparation.insert_at_end("roll")
sushi_preparation.insert_at_beginning("assemble")

print("BEFORE:")
sushi_preparation.display()

sushi_preparation.remove_beginning()
sushi_preparation.remove_at_end()
sushi_preparation.remove_at(None)

print("AFTER:")
sushi_preparation.display()

```
